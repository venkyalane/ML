```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv(r'C:\Users\MOHAN\ML\train_data.csv')
```

```python
df
```

```python
sns.lmplot(x='age',y='premium',data=df)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['age']],df['premium'])
```

```python
reg.predict([[21]])
```

```python
reg.predict([[50]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#validating our model
#linear equation: y = mx + c
#ex. premium = m * age + c
age = int(input("Enter age for finding premium: ")
premium_age_21 = reg.coef_ * age + reg.intercept_
print(f"premium_of_age_{age}: ",premium_age_21)
```
